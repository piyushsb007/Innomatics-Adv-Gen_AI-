# Task 1: Conceptual Understanding 

## Q1. What is the difference between "Love" and "love" in NLP?
### Answer
In NLP text is processed as tokens, and words with different letter cases are treated as different tokens.
"Love" and "love" are treated as different tokens because text processing systems are case-sensitive. 
However, during preprocessing, text is usually converted to lowercase, making both words identical. 
This normalization reduces vocabulary size and improves computational efficiency.

## Q2. What happens if stopwords are not removed?
### Answer
If stopwords are not removed, vocabulary size increases, leading to higher computational cost, noise in model. Common words like dominate the analysis, reducing the model ability to focus on meaningful words, which negatively impact in the performance in tasks like classification and sentimental analysis.

## Q3. Provide two real-world scenarios where removing stopwords can be harmful.
### Answer
**Scenarios :-**
#### 1. Sentimental anyalsis / Opinion making:
A movie review classifier trained without stopwords misclassifies negative opinions as positive, causing low accuracy.

For Example -
Sentence: "I do not love this movie"

With stopwords removed: "love movie" -> Positive sentiment (WRONG)

With stopwords kept: "not love movie" -> Negative sentiment (CORRECT)
#### 2. Question answering:
In question-answering systems, stopwords like 'what', 'how', 'is' define the question type needed, so removing them gives wrong answers.

For Example -
Question: "What is machine learning?"

With stopwords removed: "machine learning" -> loses the question structure.

With stopwords kept: "What is machine learning" → model understands it's a definition request.

## Q4. Explain the difference between stemming and lemmatization.
### Answer
Stemming is a simple rule-based method that cuts prefixes/suffixes, often producing incorrect words like 'running' -> 'runn'. 
Lemmatization is more accurate because it understands word meaning and converts to proper base forms, like 'running' -> 'run', 'went' -> 'go', 'better' -> 'good'.
However, stemming is faster and lemmatization is more accurate, so stemming is used for speed tasks, while lemmatization is used when accuracy matters.

# Task 2:Build Advanced Preprocessing Function


```python
import re
import string
import nltk

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english')) # unique stopwords
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
    
    # Lowercasing
    text = text.lower()

    # Remove URLs & Email
    text = re.sub(r'http\S+|www\.\S+','',text)
    text = re.sub(r'\S+@\S+' , '' , text)

    # Remove numbers
    text = re.sub(r'\d+','', text)

    # Handles repeated Charaters (soooo -> so)
    text = re.sub(r'(.)\1{2,}' , r'\1' , text)
    
    # Remove Special characters/Emoji 
    text = re.sub(r'[^\w\s]','', text)
    
    # Tokenization
    tokens = text.split()
    
    # Removes stopwords and short words(len <= 2 ).Except 'no','not'  
    # tokens = [
    #     word for word in tokens
    #     if word in ['no','not'] or (word not in stop_words and len(word)>2)
    # ]
    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]
    
    # Lemmatization
    # tokens = [lemmatizer.lemmatize(word) for word in tokens]

    # Recreate cleaned sentence
    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence

# Example test case 1
example = "WOW!!! This is soooo coool!!! Visit http://example.com now"
tokens, cleaned = preprocess_text(example)

print(f"Original: {example} \nTokens:", tokens)
print("Cleaned sentence:", cleaned)
print("-"*70)
# Example test case 2
example = "This is good" 
tokens, cleaned = preprocess_text(example)

print(f"Original: {example} \nTokens:", tokens)
print("Cleaned sentence:", cleaned)
```

# Task 4: Token Analytics


For each processed sentence, compute: 
1. Total number of tokens 
2. Number of unique tokens 
3. Average token length 

```python
def token_analysis(tokens):
    total = len(tokens)
    unique = len(set(tokens))
    avg_len_token = round(sum(len(word) for word in tokens) / total if total > 0 else 0 , 2)

    return total,unique,avg_len_token

print("======== Token Analysis ========")

for sentence in test_sentences:
    tokens , cleaned = preprocess_text(sentence)
    total,unique,avg_len_token = token_analysis(tokens)

    print("\nOriginal Sentence: ",sentence)
    print("Cleaned Sentence: ",cleaned)
    print("Total Tokens: ", total)
    print("Unique Tokens: ",unique)
    print("Average Token Length",avg_len_token)
    print("-"*80,end="")
```

```python
def full_pipeline(text_list):
    # Error handling 
    if not isinstance(text_list,list):
        raise ValueError("Input must be a list of strings")

    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens,clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens" : all_tokens,
        "clean_sentences" : clean_sentences
    }

# Example test 
pipeline_output = full_pipeline(test_sentences)

print("All Tokens:\n", pipeline_output["tokens"])
print("\nClean Sentences:\n", pipeline_output["clean_sentences"])
```
