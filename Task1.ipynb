## 1. User Login Check
** Problem Statement: **
Given a username and password, check whether login is successful.
username = "admin"
password = "1234"

Requirements:
Print "Login Successful" if both username and password match
Otherwise print "Invalid Credentials"
Real-World Application: Authentication systems

```python
print("1. Real-World Application: Authentication systems")
username = input("Enter Username: ")
password = input("Enter Password: ")

if(username=="admin" and password=="1234"):
    print("Login Successful")
else:
    print("Invalid Credentials")
```

## 2. Pass / Fail Analyzer
** Problem Statement: **
Analyze student results.
marks = [45, 78, 90, 33, 60]

Requirements:
A student passes if marks â‰¥ 50
Count the total number of pass students
Count the total number of fail students
Print the final result clearly
Real-World Application: Academic evaluation systems


```python
print("2. Real-World Application: Academic evaluation systems")
marks = [45, 78, 90, 33, 60]
count_p = 0
count_f = 0
for i in marks:
    if (i>=50):
        count_p+=1 #take count of pass std
    else:
        count_f+=1 #take count of fail std

print("------Final Result------")  
print("Total number of pass students",count_p)
print("Total number of fail students",count_f)
```

## 3. Simple Data Cleaner
** Problem Statement: **
Clean and standardize user names.
names = [" Alice ", "bob", " CHARLIE "]

Requirements:
Remove extra spaces
Convert all names to lowercase
Print the cleaned list
Real-World Application: Data preprocessing before analysis


```python
print("3. Real-World Application: Data preprocessing before analysis")
names = [" Alice ", "bob", " CHARLIE "]
list = []
for i in names:
    a = i.strip() # Remove extra spaces
    b = a.lower() # Convert all names to lowercase
    list.append(b)
# Print the cleaned list
print(list)
```

## 4. Message Length Analyzer
** Problem Statement: **
Analyze message sizes.
messages = ["Hi", "Welcome to the platform", "OK"]

Requirements:
Print the length of each message
Flag messages longer than 10 characters
Real-World Application: Text filtering and validation systems

```python
print("4. Real-World Application: Text filtering and validation systems")
messages = ["Hi", "Welcome to the platform", "OK"]

for i in range(len(messages)):
    length = len(messages[i])
    if(length>10):
        print(f"Message longer than 10 characters with length {length}")
    else:
        print(f"Length of msg is {length}")
```

## 5. Error Message Detector
** Problem Statement: **
Detect error messages from system logs.
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

Requirements:
Count the number of "ERROR" entries
Print the total error count
Real-World Application: Monitoring and log analysis systems


```python
print("5. Real-World Application: Monitoring and log analysis systems")
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
count=0
for i in logs:
    if i=="ERROR":
        count+=1
print("Total number of \"ERROR\" entries are",count)
```
