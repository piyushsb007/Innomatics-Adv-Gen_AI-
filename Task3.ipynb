# Problem Statement 1: Employee Performance Bonus Eligibility
## Description:
A company evaluates employee performance scores at the end of the year. You are given a dictionary containing employee names and their performance scores.


```python
print("Problem Statement 1: Employee Performance Bonus Eligibility")
employees = {"Ravi": 92,"Anita": 88,"Kiran": 92,"Suresh": 85}
highest_score = max(employees.values()) # to get highest performance score 

# To get employees name with highest score
top_performers = [] 
for name in employees:
    if employees[name] == highest_score :
        top_performers.append(name)
# Result 
print("Top performers Eligible for Bonus:",end=" ")

for i in range(len(top_performers)):
    if i != len(top_performers) - 1 :
        print(top_performers[i],end=", ")
    else:
        print(top_performers[i],end=" ")
print("(Score:",highest_score, ")")        
```

# Problem Statement 2: Search Query Keyword Analysis
## Description:
 An e-commerce website stores customer search queries. You are given a search query sentence entered by a user.


```python
print("Problem Statement 2: Search Query Keyword Analysis")
search_query = "Buy mobile phone buy phone online"
import string
# Remove punctuation
translation_table = str.maketrans("","",string.punctuation)
search_query = search_query.translate(translation_table)
# Convert to lowercase and split into words
words = search_query.lower().split()
# Frequence Count
freq = {}

for word in words: 
    freq[word] = freq.get(word,0) + 1
result = {}
for key,value in freq.items():
    if value > 1:
        result[key] = value
print(result)
```

# Problem Statement 3: Sensor Data Validation
## Description:
A factory collects sensor readings every hour. Each reading is stored in a list where the index represents the hour and the value represents the sensor reading.


```python
print("Problem Statement 3: Sensor Data Validation ")
sensor_readings = [3, 4, 7, 8, 10, 12, 5] # Input readings
valid_readings = [] # Empty to store valid readings
n = len(sensor_readings) # Length of input reading list
for hour_index in range(n):
    value = sensor_readings[hour_index]
    if value%2 == 0:
        valid_readings.append((hour_index,value))
print("Valid Sensor Readings (Hour, Value): \n",valid_readings)
```

# Problem Statement 4: Email Domain Usage Analysis
## Description:
A company wants to analyze which email providers its users are using. You are given a list of employee email IDs.

```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
] 
email_count = {}
for email in emails:
    email = email.split("@")[1]
    email_count[email] = email_count.get(email,0) + 1

total = len(emails)

for key , value in email_count.items():
    percentage = value/total * 100
    print(f"{key}: {percentage:.0f}%")
```

# Problem Statement 5: Sales Spike Detection
## Description:
A retail company tracks daily sales. Sudden spikes in sales may indicate promotions or unusual activity.


```python
print("Problem Statement 5: Sales Spike Detection")
sales = [1200, 1500, 900, 2200, 1400, 3000]
# Calculate average 
avg_daily_sales = sum(sales)/len(sales)
# Days where sales are more than 30% above average.
threshold = avg_daily_sales + (avg_daily_sales)*0.3
print("Threshold: ",threshold)
for day in range(len(sales)):
    if sales[day] > threshold:
        print(f"Day{day+1}:",sales[day])
```

# Problem Statement 6: Duplicate User ID Detection
## Description:
A system stores user IDs during registration. Duplicate IDs can cause data integrity issues.


```python
print("Problem Statement 6: Duplicate User ID Detection")
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
unique_user = {} # Dict to store unique users

# loop to count appearance of a each user
for user in user_ids:
    unique_user[user] = unique_user.get(user,0) + 1 
for user,appearance in unique_user.items(): # to print required users
    if user != "user2":
        print(f"{user} → {appearance} times")
```
