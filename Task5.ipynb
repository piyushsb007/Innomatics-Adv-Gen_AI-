# Problem 1: Social Media – Post Engagement Analyzer 
## Problem Statement 
Create a function to analyze engagement on a social media post. 

Rules 
* Calculate total likes 
* If Total Likes ≥ 1000 → Viral Post 
* Otherwise → Normal Engagement

```python
def post_analyzer(like_list):
    total_likes = 0
    #Calculate likes
    for likes in like_list:
        total_likes += likes
    # Check engagement of post
    if total_likes >= 1000:
        return f"Total likes: {total_likes}\nPost Status: Viral Post "
    else:
        return f"Total likes: {total_likes}\nPost Status: Normal Engagement "

# For Example
# list containing, Likes across various platforms
likes_list = [200,230,250,100,200] 
# function call 
print(post_analyzer(likes_list)) #Normal engagement 

likes_list = [207,230,250,150,200]
print(post_analyzer(likes_list)) #Viral
```

# Problem 3: Agriculture – Rainfall Adequacy Checker
## Problem Statement
Create a function that analyzes seasonal rainfall data.

    Rules
* Calculate average rainfall
* Average ≥ required level → Adequate Rainfall
* Otherwise → Inadequate Rainfall


```python
def rainfall_checker():
    n = int(input("Enter number of days/seasons: "))
    rainfall = []
    total_r = 0    
    # loop to get rainfall input
    for i in range(n):
        r = float(input(f"Enter rainfall for a day {i+1}: "))
        rainfall.append(r)
        total_r += r
     # Calculate Avg Rainfall   
    avg_rain = total_r/n
    required_level = float(int(input("Enter required rainfall level: ")))
    print("Average Rainfall: ",avg_rain)
    # Check condition
    if avg_rain >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")
rainfall_checker()
```

# Problem 4: Social Media – Duplicate Account Detection
## Problem Statement
Create a function to detect duplicate usernames in a system.

Rules
* If duplicates exist → Flag accounts
* Else → Accounts are unique


```python
# Duplicate Account Detection function
def socialmedia_acc(acc_list):
    # Check duplicates
    acc_list = [i.lower() for i in acc_list] 
    # used set to get only unique username
    unique_acc = set(acc_list) 
    if len(acc_list) != len(unique_acc):
        return "Yes" # Flag account
    else:
        return "No" # Accounts are unique
usernames = ["Piyush","sathya","piyush","raj","ved"]
print("Duplicate Accounts Found: ",socialmedia_acc(usernames))
```

# Problem 6: Agriculture – Premium Crop Price Filter
## Problem Statement
Create a function that filters crop prices above a premium threshold.

Rules
* Price > ₹2000 → Premium Crop
* Else → Regular Crop


```python
# function to filter crop prices into Premium and regular
def crop_price_filter(crops):
    threshold = 2000  # Premium price threshold
    premium_crop = []
    regular_crop = []
    # Loop through each crop price in the list
    for price in crops:
        if price > threshold:
            premium_crop.append(price) # Premium crops
        else:
            regular_crop.append(price) # Regular crops
    return premium_crop, regular_crop

# Input number of crop prices
n = int(input("Enter the number of Crops price to Filter: "))
crops = [] #list to store crop prices
for i in range(n):
    crops.append(int(input(f"Enter Crop{i+1} : "))) 
# function call
premium_crop, regular_crop = crop_price_filter(crops)
# Display result
print(f"Premium Crops: {premium_crop}")
print(f"Regular Crops: {regular_crop}")
# Premium Crops: [2500, 3200]
```

# Problem 7: System Monitoring – Application Health Checker
## Problem Statement
Create a function to evaluate application health based on error count.

Rules
* Errors = 0 → Healthy
* Errors ≤ 5 → Minor Issues
* Errors > 5 → Critical Issues

```python
def health_monitoring():
    error_count = int(input("Enter the Error Count: "))
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else: # error_count > 5
        status = "Critical Issues"
    return error_count , status
error_count , status = health_monitoring()
print(f"Error Count: {error_count} \nSystem Status: {status}")
```
