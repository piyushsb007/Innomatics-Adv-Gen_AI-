# Problem 1: Unique Words in a Sentence
## Problem Statement: Given a sentence, identify all unique words and display: 
### 1.The count of unique words. 
### 2.The unique words themselves


```python
Sentence = input("Enter sentence :") # Taking input from users
words = Sentence.split() # Extracting words from sentence
new_senlist = [] # new sentence list to store words from sentence
count = 0 #var to store count
for word in words:
    if word not in new_senlist: # Checking for duplicates
         new_senlist.append(word)# Appending unique word to new sentence list
         count+=1
print("Unique words count: ",count) 
print("Unique words:" ,new_senlist)
```

# Problem 2: Highest Salary from Employee Data
## Problem Statement:Store employee names and their salaries, and determine which employee has the highest salary.


```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
max_value= 0
max_key = None
for name, salary in employees.items():
    if (salary > max_value):
        max_value = salary
        max_key = name
print("Highest Salary:",max_key,"-",max_value)
```

# Problem 4: Count Products Above a Price Threshold
##Problem Statemen: Count how many products have a price greater than 1000.

```python
prices = [450, 1200, 899, 1500, 300]
def product(prices):
    count = 0 
    for price in prices:
        if price>1000:
            count+=1
    print("Product above 1000: ",count)
product(prices)
```

# Problem 5: Calculate Attendance Percentage
## Problem Statement: Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).

```python
attendance = ["P", "P", "A", "P", "P"]
count=0
total_days = len(attendance)
for i in attendance:
    if i == "P":
        count+=1
perct = (count/total_days)*100
print("Attendance Percentage:",perct)
```

# Problem 10: Calculate Average Marks
## Problem Statement: Calculate the average marks of a student from a list of marks.

```python
def avg(marks):
    sum = 0
    for m in marks:
        sum += m
    total_subject = len(marks) #total no. of subjects
    avg_marks = (sum / total_subject)
    return avg_marks

marks = [87,78,90,60]
print("Average marks of students: ",avg(marks))
```
