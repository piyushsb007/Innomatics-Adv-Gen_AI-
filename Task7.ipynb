# Problem 1: Smart Light Controller
## Problem Statement
Create a class to control a smart light that can be turned ON or OFF and display its status.
🌍 Real-Time Use
Smart home automation systems
*  Sample Input
Light Name: Bedroom Light
Action: ON
* Expected Output
Bedroom Light is ON

```python
class SmartLight:
    def __init__(self,light_name):
        self.light_name = light_name
        self.status = "OFF" # Default status
    def turn_on(self):
        self.status = "ON"
    def turn_off(self):
        self.status = "OFF"
    def display_status(self):
        print(self.light_name,"is",self.status)

# Input
light_name = input("Light Name: ")
action = input("Action: ")

# Creating object
light = SmartLight(light_name)

#Performing action
if action.lower() == "on":
    light.turn_on()
elif action.lower() == "off":
    light.turn_off()
else:
    print("Invalid action! Light remains OFF")

#Diplay final result
light.display_status()
```

# Problem 2: Employee ID Card System
## Problem Statement
Create a class that stores employee name, ID, and department and displays ID card details.
### Real-Time Use
Corporate HR systems
* Sample Input

Employee Name: Rahul

Employee ID: EMP102

Department: IT
* Expected Output

Employee ID Card

Name: Rahul

ID: EMP102

Department: IT


```python
class Employee:
    def __init__(self,name,emp_id,dept):
        self.name = name
        self.emp_id = emp_id
        self.dept = dept 
    def display_id_card(self):
        print("\nEmployee ID Card ")
        print("Name: ",self.name)
        print("ID: ",self.emp_id)
        print("Department: ",self.dept)
# Taking input
name = input("Employee Name: ")
emp_id = input("Employee ID: ")
dept = input("Department: ")
# Creating object
emp = Employee(name,emp_id,dept)
# Validation
if name.strip() and emp_id.strip() and dept.strip():
    emp = Employee(name, emp_id, dept)
    # Diplay emp info
    emp.display_id_card()
else:
    print("Invalid input! Fields cannot be empty.")
```

# Problem 3: Mobile Contact Record
## Problem Statement
Create a class to store contact name and phone number and display contact information.

* Real-Time Use
** Mobile contact applications

## Sample Input
* Contact Name: Anita
* Phone Number: 9876543210
## Expected Output
* Contact Saved
* Name: Anita
* Phone: 9876543210


```python
class Contact:
    def __init__(self,name,phone_no):
        self.name = name 
        self.phone_no = phone_no
    def display_contact_info(self):
        print("\nContact Saved")
        print("Name:",self.name)
        print("Phone:",self.phone_no)
# Taking Input
name = input("Contact Name: ")
phone_no = input("Phone Number: ")
#creating object 
contact = Contact(name,phone_no)
# Validation
if phone_no.isdigit() and len(phone_no) == 10:
    contact = Contact(name, phone_no)
    # Diplay Saved Contacts
    contact.display_contact_info()
else:
    print("Invalid phone number! Enter a 10-digit number.")

```

# Problem 4: Product Price Tag Generator
## Problem Statement
Create a class that stores product name and price and prints a formatted price tag.
 
### Real-Time Use
Retail billing systems
* Sample Input:
  
Product Name: Headphones

Price: 2499

* Expected Output:
  
Product: Headphones

Price: ₹2499


```python
class Product:
    def __init__(self,name,price):
        self.name = name
        self.price =price
    def display_tag(self):
        print("\nProduct:",self.name)
        print("Price:₹",self.price)
# Taking Input
name = input("Product Name:")
price = input("Price:")

# Creating object 
product = Product(name,price)

# Display 
product.display_tag()
```

# Problem 5: Movie Rating Display System
## Problem Statement
Create a class that stores movie name and rating and displays movie details.
### Real-Time Use
Streaming platforms
# Sample Input
* Movie Name: Inception
* Rating: 4.8
# Expected Output
* Movie: Inception
* Rating: 4.8 / 5


```python
class Movie:
    def __init__(self,name,rating):
        self.name = name
        self.rating = rating
    def display_movie_details(self):
        print("\nMovie:",self.name)
        print(f"Rating: {self.rating:.1f} / 5")
# Taking Input
name = input("Movie Name:")
rating = float(input("Rating:"))
# Creating object
obj = Movie(name,rating)
# Display Movie Rating
obj.display_movie_details()
```

# Problem 6: Delivery Address Manager
# Problem Statement
Create a class that stores customer name and delivery address and prints delivery details.
### Real-Time Use
Courier and logistics platforms
# Sample Input
* Customer Name: Suman
* Address: Hyderabad
# Expected Output
* Delivery Details
* Customer: Suman
* Address: Hyderabad


```python
class Delivery:
    def __init__(self,name,address):
        self.name = name
        self.address = address
    def display_details(self):
        print("\nDelivery Details")
        print("Customer:",self.name)
        print("Address:",self.address)
# Taking input 
customer_name = input("Customer Name:")
address = input("Address:")
# Creating object 
delivery = Delivery(customer_name,address)
# Display Delivery details
delivery.display_details()
```
