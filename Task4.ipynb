# Problem 1: Mobile Recharge Validation System
## Problem Statement : Create a function to validate a mobile recharge request.
### Rules:
* Recharge amount must be ≥ ₹50
* Valid plans: 199, 299, 399, 599
* User should be allowed to retry if invalid


```python
def recharge_valid():
    while(True):
        recharge_amount = int(input("Enter Recharge Amount(Valid plans: 199, 299, 399  599): "))
        if recharge_amount < 50:
           print("Minimum recharge amount is ₹50. Retry.\n")
        elif recharge_amount in [199,299,399,599]:
            print(f"Recharge of ₹{recharge_amount} is Successfull")
            break
        else:  
            print("Invalid(Choose Valid plans: 199, 299, 399  599): Retry")

recharge_valid()
```

# Problem 2: Inventory Reorder Alert System
## Problem Statement
Given a dictionary of products and stock quantities:
* Print Reorder Alert if stock < 15
* Print Stock OK otherwise

```python
 rec = {"Shirt":20,"pant":23,"Jacket":21,"Football":9}
def inventory_alert():
    for key, value in rec.items():
        if value < 15:
            print(f"Reorder Alert {key}")
        else:
            print(f"{key} stock OK")

inventory_alert()
```

 # Problem 3: Student Result Processing System
## Problem Statement
Create a function that:
* Accepts a list of student marks
* Calculates average marks
* Prints Pass if average ≥ 50 else Fail

```python
def std_result(marks_list):
    total = 0 
    # loop to calculate marks
    for m in marks_list:
        total += m
    # calculate avg marks of student
    avg_marks = total / len(marks_list)
    print("Average Marks: ",avg_marks)
    # Check pass or fail
    if avg_marks>50:
        return "Pass"
    else:
        return "Fail"
# Example 
student_marks1 = [40,32,38,29,84]   
student_marks2 = [90,52,83,29,34] 
print(std_result(student_marks1))
print(std_result(student_marks2))
```

# Problem 4: Cab Fare Estimator with Retry Option
## Problem Statement
Create a cab fare calculator:
* Base fare = ₹50
* ₹12 per km
* Peak hour → 25% extra
* Allow retry using while loop


```python
def fare_estimator():
    distance = int(input("Enter distance to be travelled(in km): "))
    while(distance > 0):
        hr = input("Searching result in Peak Hour [P] /Regular Hour [R]: ")
        base_fare = 50
        cost = base_fare + (distance*12)
        if hr.lower()=="p":
            cost += cost*0.25
            print(f"Total fare cost in Peak Hour is ₹{cost} and travelled {distance}km")
            break
        elif hr.lower() == "r":
            print(f"Total fare cost is ₹{cost} and travelled {distance}km")
            break
        else:
            print("Invalid input")

fare_estimator()
```

```python
def calculate_fare(distance,hr):
    base_fare = 50
    per_km = 12
    #normal hour fare cost
    fare_cost =  base_fare + (distance*per_km)
    # peak hour fare cost
    if hr.lower() == "y":
        fare_cost += fare_cost*0.25  
    return fare_cost

while True:
    distance = int(input("Enter distance (in km): "))
    if distance <= 0:
        print("Distance must be greater than 0")
        continue
    hr = input("Is it peak hour? [y/n]: ")
    print(f"Total fare : {calculate_fare(distance,hr):.2f}")

    retry = input("Do you want to calculate again? (yes/no): ")
    if retry.lower() != "yes":
        print("Thank you! Use Agian")
        break
        
```

# Problem 5: Employee Attendance Eligibility Checker
## Problem Statement
Create a function that:
* Accepts attendance list ("P", "A")
* Calculates attendance percentage
* Returns Eligible if ≥ 75% else Not Eligible


```python
def attendance_check(attendance):
    total_p = 0
    #Calculate attendance percentage
    for i in attendance:
        if i.lower()=="p":
            total_p += 1
    attendance_per = (total_p /len(attend))*100
    print("Employee Attendance is",attendance_per , end =" ")
    if attendance_per >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
# Examples
attendance_list1 = ["P","A","P","P","A","A","P","A"]
attendance_list2 = ["P", "P", "A", "P", "P", "A", "P", "P"]
print(attendance_check(attendance_list1))
print(attendance_check(attendance_list2))
```

# Problem 6: Password Strength Checker
## Problem Statement
Create a function to check password strength.

** Rules: **
* Minimum length = 8
* Must contain at least one digit
* Must contain at least one special character (@#$)


```python
def check_password_strength(password):
    # Minimum length = 8
    if len(password) < 8:
        return "Weak Password (Length < 8)"
    # Must contain at least one digit
    digit = False
    special_char = False
    for char in password:
        if char.isdigit():
            digit = True
        if char in "@#$":
            special_char = True
            
    if digit and special_char:
        return "Strong Password"
    else: 
        # Must contain at least one special character (@#$)
        return "Weak Password (Must contain atleast one digit and one special character)"

Password = input("Enter Password to Check password strength: ")
result = check_password_strength(Password)
print(result)
```
